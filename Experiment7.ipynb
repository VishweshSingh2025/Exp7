```python
#3 Factorial of  no.
n=int(input("Enter a number:"))
fact=1
while n>0:
   fact*=n
   n-=1
print('factorial =',fact)
```

```python
#4 Fibonacci sequence
n=int(input("Enter the number of terms:"))
a=0
i=1
terms=1
while terms<=n:
  print(a,end=' ')
  c=a+i
  a=i
  i=c
  terms+=1
```

```python
#5 Fibonacci sequence with limits
limit=int(input("Enter the limit:"))
a=0
i=1
while a<=limit:
  print(a,end=' ')
  a , i=i, a+i
```

```python
#6 Reverse of a number
n=int(input("Enter a number:"))
rev=0
while n>0:
  digit = n % 10
  rev = rev*10 + digit
  n //= 10
print('reversed no. : ', rev)
```

```python
#7 Palindrome or Not
n=int(input("Enter a number:"))
temp=n
rev=0
while n>0:
  digit = n % 10
  rev = rev*10 + digit
  n //= 10
print('reversed no. : ', rev)
if temp==rev:
  print('Palindrome')
else:
  print('Not a Palindrome')
```

```python
#8 Palindrome of a String
s=input("Enter a string:")
i,j=0,len(s)-1
while i<j:
  if s[i]!=s[j]:
    print('Not a Palindrome')
    break
    i+=1
    j-=1
  else:
    print('Palindrome')
    break
```

```python
#9 Check palindrom of a string
s=input("Enter a string:")
rev=s[::-1]
if s==rev:
  print('Palindrome')
else:
  print('Not a Palindrome')
```

```python
#10 Count Digits in a Number
num=int(input("Enter a number:"))
count=0
while num>0:
  num//=10
  count+=1
print(count)
```

```python
#11 Search an element in a list
nums=[10,20,30,40,50]
key=int(input("Enter a number:"))
i=0
while i<len(nums):
  if nums[i]==key:
    print('Element is found at index',i)
    break
  i+=1
else:
    print('Element not found')
```
